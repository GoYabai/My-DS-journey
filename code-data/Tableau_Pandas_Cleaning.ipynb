# I. Import dataset

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/Sample Superstore - Region Localization - Dirty.csv")
```

# II. Understanding dataset

```python
# Hiển thị các thông tin cơ bản
print(f"Số dòng: {len(df)}")
print(f"Số cột: {len(df.columns)}")
df.head()
```

# III. Detecting Data Quality Problems

```python
# Kiểm tra thông tin các cột
df.info()
```

```python
df.describe()
```

## Missing Values

```python
# Kiểm tra giá trị null
null_counts = df.isnull().sum()
null_counts[null_counts > 0]
```

Kết quả phát hiện: Có 21 giá trị null trong Customer Name và 13 giá trị null trong Product Name. Đối với dữ liệu bán hàng, việc thiếu thông tin này có thể ảnh hưởng đến phân tích. Vì tỉ lệ null trong dataset không nhiều nên ta có thể drop các dòng này.

```python
# Loại bỏ các dòng có giá trị null trong các cột quan trọng
df_cleaned = df.dropna(subset=['Customer Name', 'Product Name'])

print(f"Số dòng sau khi loại bỏ null: {len(df_cleaned)}")
print(f"Đã loại bỏ: {len(df) - len(df_cleaned)} dòng")
```

## Duplicates Values

```python
# Kiểm tra số lượng dòng bị đúp
print(f"Số dòng bị trùng: {df_cleaned.duplicated().sum()}")
print(f"Số dòng sau khi loại bỏ trùng: {len(df_cleaned.drop_duplicates())}")
```

```python
df = df_cleaned.drop_duplicates().copy()
```

## Data Type

Đánh giá Data type hiện tại:
- Các cột có giá trị số nguyên: Order Quantity
- Các cột có giá trị số thập phân: Sales, Discount, Profit
- Các cột có giá trị text như: Province, Order ID, Customer Name, Product Name, Category, Sub-Category

Tuy nhiên **Order Date và Ship Date**: Hiện đang là object (text) cần chuyển sang datetime64.

```python
# Chuyển đổi các cột ngày tháng sang datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Kiểm tra lại
print("Sau khi chuyển đổi:")
df[['Order Date', 'Ship Date']].info()
```

# IV. Checking Business Logic

## Numeric fields

```python
numeric_cols = ['Order Quantity','Sales','Profit']

for col in numeric_cols:
  print(df[col].describe())
```

```python
# Kiểm tra giá trị âm
negative_qty = df_cleaned[df_cleaned['Order Quantity'] < 0]
print(f"Số dòng có Order Quantity âm: {len(negative_qty)}")
print(negative_qty.value_counts())

```

Số lượng đơn hàng không thể âm. Các giá trị âm có thể là lỗi nhập liệu, ta chuyển về giá trị dương.

```python
# Chuyển giá trị âm thành giá trị tuyệt đối
df_cleaned['Order Quantity'] = df_cleaned['Order Quantity'].abs()
```

## Text fields

```python
# Kiểm tra giá trị của các cột category
text_cols = ['Shipping Service','Segment','Category','Sub-Category']

for col in text_cols:
  print(df[col].value_counts())
```

Có sự trùng lặp giữa các phân loại về dịch vụ giao hàng tại cột Shipping Service:
- "Same Day" và "Same-Day"
- "1st class" và "First Class"
- "2nd class" và "Second Class"
- "Std class" và "Standard Class"

```python
# Chuẩn hóa tên dịch vụ giao hàng
shipping_mapping = {
    'Same-Day': 'Same Day',
    '1st class': 'First Class',
    '2nd class': 'Second Class',
    'Std class': 'Standard Class'
}

df['Shipping Service'] = df['Shipping Service'].replace(shipping_mapping)
```

```python
# Kiểm tra giá trị của các cột địa lý
geo_cols = ['Country','Province']
pd.set_option('display.max_rows', None)
for col in geo_cols:
  print(df[col].value_counts())
```

Giá trị quốc gia chỉ có 1 giá trị là Việt Nam, tuy nhiên với các tỉnh thành có sự trùng lặp.
- "Ha Noi" và "HN"
- "Tp.HCM" và "Ho chi Minh"
- Thua Thien Hue" và "Hue"

```python
# Chuẩn hóa tên tỉnh/thành phố
province_mapping = {
    'HN': 'Ha Noi',
    'Ho chi Minh': 'Tp.HCM',
    'Hue': 'Thua Thien Hue'
}

df['Province'] = df['Province'].replace(province_mapping)
```

## Date fields

```python
date_cols = ['Order Date','Ship Date']

for col in date_cols:
  print(df[col].describe())
```

Thời gian của Ship Date diễn ra từ đầu năm 2014 đến hết năm 2017, trong khi đó thời gian của Order Date lại đang xuất hiện thông tin bất thường là năm 1899.

```python
df['Order Date'].dt.year.value_counts().sort_index()
```

Số lượng dòng có bất thường về năm của Order Date là 26 dòng, đây là một con số nhỏ với dataset của chúng ta, vì vậy có thể drop. Tuy nhiên nếu muốn điền vào bằng một giá trị năm phù hợp ta có thể tham khảo code sau.

```python
# Kiểm tra phân phối của số ngày từ lúc đặt đến khi giao hàng
(df['Ship Date'] - df['Order Date']).dt.days.describe()
```

```python
# Kiểm tra và sửa Order Date năm 1899 với thông tin Order Date = Ship Date - 4 ngày
order_1899 = df['Order Date'].dt.year == 1899

lead_time = (
    df.loc[~order_1899, 'Ship Date']
    - df.loc[~order_1899, 'Order Date']).median()

print(f"Thời gian lead time trung bình là: {lead_time}")

df.loc[order_1899, 'Order Date'] = (
    df.loc[order_1899, 'Ship Date'] - lead_time)
```

# Kiểm tra kết quả cuối cùng

```python
# Thông tin tổng quan
print(f"Số dòng: {len(df)}")
print(f"Số cột: {len(df.columns)}")

# Kiểm tra lại giá trị null
null_check = df.isnull().sum()
print(null_check[null_check > 0] if null_check.sum() > 0 else "Không còn giá trị null")
```
